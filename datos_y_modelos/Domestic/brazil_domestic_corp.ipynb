### 1) Leer el archivo Excel

```python
import pandas as pd
import numpy as np
from scipy.stats import kurtosis

file_path = 'brazil_domestic_corp_db.xlsx'

df = pd.read_excel(file_path, sheet_name='db_values_only')

df = df.replace(r'^#N/A.*', np.nan, regex=True)

df = df.dropna(how='all')

df
```

```python
df.to_excel('1.xlsx', index=False)
```

### 1.1) Mostrar las columnas

```python
df.columns
```

### 2) en "CLASSIFICATION_LEVEL_4_NAME" quitar los que son estatales (estados varios)

```python
df = df[~df['CLASSIFICATION_LEVEL_4_NAME'].str.startswith("Government", na=False)]
df
# df.to_excel('2.xlsx', index=False)
```

### 2.1) en "industry_sector" quitar los que son financeiros (eg. banks)

```python
df = df[~df['industry_sector'].isin(['Financial'])]
df
```

### 2.2) en "CPN_TYP" filtrar por "FIXED" (quitar los que son "FLOATING", "DEFAULT", "FLAT TRADING", "VARIALE" y "WHEN ISSUED")

```python
df = df[df['CPN_TYP'].isin(['FIXED'])]
df
```

### 2.3) en "MTY_TYP" filtrar por "AT MATURITY" (quitar los que son "CALL/SINK", "CALLABLE", "CONVERTIBLE", "EXTENDIBLE", "SINK/EXT" y "SINKABLE")

```python
df = df[df['MTY_TYP'].isin(['AT MATURITY'])]
df
```

### 2.4) en "CRNCY" filtrar por "BRL" (quitar los que son denominados en moneda extranjera)

```python
df = df[df['CRNCY'].isin(['BRL'])]
df
```

### 2.5) en "TOT_DEBT_TO_EBITDA", se excluyeron aquellos valores que comienzan con "#N/A" ya que la falta de datos fundamentales indica que dichos emisores no cotizan en bolsa.

```python
df['TOT_DEBT_TO_EBITDA'] = pd.to_numeric(df['TOT_DEBT_TO_EBITDA'], errors='coerce')
df = df[df['TOT_DEBT_TO_EBITDA'].notna()]
df
```

En contraste, los bonos con "N" en la muestra son 23 bonos pre-fijados que pagan intereses nominales (principal no se ajusta por inflaci√≥n). Para poder comparar de manera adecuada, con los bonos vinculados a la inflaci√≥n (69 bonos), ser√≠a necesario restar la tasa de inflaci√≥n del cup√≥n para igualar las bases. Dato que este calculo no es tan sencillo, por el momento, estos bonos pre-fijados (23) se mantendr√°n fuera de la muestra.

```python
df = df[df['INFLATION_LINKED_INDICATOR'].isin(['Y'])]
df
```

### Funci√≥n de Estad√≠sticas descriptivas sobre la muestra filtrada

```python
def descriptive_stats(col):
    return {
        'mean': col.mean(),
        'min': col.min(),
        'max': col.max(),
        'median': col.median(),
        'std_dev': col.std(),
        'kurtosis': kurtosis(col.dropna())
    }
```

### 4) luego agrupar por empresa los bonos que emiti√≥ cada uno y ahi vemos cuantos hay de cada una , con que maturity, volume issued y sobre todo volume traded as√≠ elegimos los m√°s l√≠quidos

```python
grouped = df.groupby(by=['ISSUER'])

issuances_count = grouped[["ID_ISIN"]].count().rename(columns={"ID_ISIN": "NUMBER_OF_ISSUANCES"})

maturity_min = grouped[["MATURITY"]].min().rename(columns={"MATURITY": "MATURITY_MIN"})
maturity_max = grouped[["MATURITY"]].max().rename(columns={"MATURITY": "MATURITY_MAX"})

volume_issued_sum = grouped[["AMT_ISSUED"]].sum().rename(columns={"AMT_ISSUED": "VOLUME_ISSUED_SUM"})
volume_issued_min = grouped[["AMT_ISSUED"]].min().rename(columns={"AMT_ISSUED": "VOLUME_ISSUED_MIN"})
volume_issued_max = grouped[["AMT_ISSUED"]].max().rename(columns={"AMT_ISSUED": "VOLUME_ISSUED_MAX"})
volume_issued_avg = grouped[["AMT_ISSUED"]].median().rename(columns={"AMT_ISSUED": "VOLUME_ISSUED_AVG"})


number_of_trades_min = grouped[["LQA_EMPIRICAL_TRADE_COUNT"]].min().rename(columns={"LQA_EMPIRICAL_TRADE_COUNT": "NUMBER_OF_TRADEDS_MIN"})
number_of_trades_max = grouped[["LQA_EMPIRICAL_TRADE_COUNT"]].max().rename(columns={"LQA_EMPIRICAL_TRADE_COUNT": "NUMBER_OF_TRADEDS_MAX"})
number_of_trades_avg = grouped[["LQA_EMPIRICAL_TRADE_COUNT"]].median().rename(columns={"LQA_EMPIRICAL_TRADE_COUNT": "NUMBER_OF_TRADEDS_AVG"})

   

maturity_summary = pd.concat([issuances_count, 
                              maturity_min, 
                              maturity_max, 
                              volume_issued_sum, 
                              volume_issued_min, 
                              volume_issued_max, 
                              volume_issued_avg,
                              number_of_trades_min,
                              number_of_trades_max,
                              number_of_trades_avg,
                             ], axis=1).reset_index()
   


maturity_summary = maturity_summary.sort_values(by="NUMBER_OF_ISSUANCES", ascending=False)

maturity_summary
```

```python
maturity_summary.to_excel('4.xlsx', index=False)
```
