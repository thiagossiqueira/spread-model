### 1) Leer el archivo Excel

```python
import pandas as pd
import numpy as np
from scipy.stats import kurtosis

file_path = 'brazil_supranational_db.xlsx'

df = pd.read_excel(file_path, sheet_name='db_values_only')

df = df.replace(r'^#N/A.*', np.nan, regex=True)

df = df.dropna(how='all')

df
```
